```python
import pandas as pd

import requests
from bs4 import BeautifulSoup
from pprint import pprint

from itertools import chain
import time
import re

from lxml import html
from lxml.cssselect import CSSSelector
from lxml .etree import tostring

from selenium import webdriver

import json


from selenium import webdriver

import json
```

```python
def codes(url):
    response=requests.get(url)
    page=response.content
    page=BeautifulSoup(page,"html.parser")
    links=["https://www.youtube.com/" + i.get("href") for i in page.find_all("a", class_=" yt-uix-sessionlink spf-link ")]
    titles=[i.get_text() for i in page.find_all("a", class_="yt-uix-tile-link yt-ui-ellipsis yt-ui-ellipsis-2 yt-uix-sessionlink spf-link ")]
    durations=[i.get_text() for i in page.find_all("span", class_="video-time")]
    usernames=[i.get_text() for i in page.find_all("a", class_="yt-uix-sessionlink spf-link ")]
    views=[i.get_text() for i in page.find_all("li", text = re.compile("view"))]
    views_list=[]
    for i in views:
        splitted=i.split(" ")
        del splitted[1::2]
        views_list.append(splitted)
    views_final=list(chain.from_iterable(views_list))
    table=pd.DataFrame(list(zip(links,titles,durations,usernames,views_final)),columns=["Links","Titles","Durations","Usernames","Views"])
    return table
```

```python
table=codes("https://www.youtube.com/feed/trending")
table
```

```python
table.to_csv("A_D_Beautiful_Soup.csv")
```
